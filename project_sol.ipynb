```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

class SolarAnalysis:
    def __init__(self, customer_columns_file, house_load_file, generation_file, ev_load_file):
        """Initialize the analysis with file paths."""
        self.customer_ids = pd.read_csv(customer_columns_file, header=None).iloc[0].values
        self.house_load = pd.read_csv(house_load_file, header=None)
        self.generation = pd.read_csv(generation_file, header=None)
        self.ev_load = pd.read_csv(ev_load_file, header=None)
        
        # Create timestamp index
        self.create_timestamp_index()
        
    def create_timestamp_index(self):
        """Create timestamp index for the data (3 years with 30-min intervals)."""
        start_date = datetime(2019, 7, 1)  # Assuming start date
        timestamps = [start_date + timedelta(minutes=30*i) for i in range(52608)]
        
        # Set index for all dataframes
        self.house_load.index = timestamps
        self.generation.index = timestamps
        self.ev_load.index = timestamps
        
    def select_sample_houses(self, n_houses=5):
        """Select a sample of houses for analysis."""
        # Random seed
        np.random.seed(0)
        # self.selected_houses = np.random.choice(self.customer_ids, size=n_houses, replace=False)
        self.selected_houses = np.array([2, 73, 74, 109, 141])
        return self.selected_houses
    
    def calculate_metrics(self, house_id):
        """Calculate self-consumption and self-sufficiency for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate self-consumption
        self_consumed = np.minimum(house_load, generation)
        self_consumption = np.sum(self_consumed) / np.sum(generation) * 100
        
        # Calculate self-sufficiency
        self_sufficiency = np.sum(self_consumed) / np.sum(house_load) * 100
        
        return {
            'house_id': house_id,
            'self_consumption [%]': self_consumption,
            'self_sufficiency [%]': self_sufficiency,
            'total_load [kWh]': np.sum(house_load),
            'total_generation [kWh]': np.sum(generation),
            'total_self_consumed [kWh]': np.sum(self_consumed)
        }
    
    def plot_daily_profile(self, house_id):
        """Create daily profile plot for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate average daily profile
        daily_load = house_load.groupby(house_load.index.hour).mean()
        daily_generation = generation.groupby(generation.index.hour).mean()
        
        plt.figure(figsize=(10, 6))
        plt.plot(daily_load.index, daily_load.values, label='Consumption (Load + EV)')
        plt.plot(daily_generation.index, daily_generation.values, label='Generation')
        plt.title(f'Average Daily Profile - House {house_id}')
        plt.xlabel('Hour of Day')
        plt.ylabel('Power (kWh)')
        plt.legend()
        plt.grid(True)
        return plt.gcf()
    
    def create_summary_report(self, selected_houses=None):
        """Create a summary report for selected houses."""
        if selected_houses is None:
            selected_houses = self.selected_houses
            
        # Calculate metrics for all selected houses
        metrics = [self.calculate_metrics(house_id) for house_id in selected_houses]
        return pd.DataFrame(metrics)

def main():
    # Initialize analysis
    analysis = SolarAnalysis(
        'CustomerColumns.csv',
        'HouseLoad.csv',
        'Generation.csv',
        'EVLoad.csv'
    )
    
    # Select sample houses
    selected_houses = analysis.select_sample_houses(5)
    print("Selected houses for analysis:", selected_houses)
    
    # Generate summary report
    summary_df = analysis.create_summary_report()
    print("\nSummary Report:")
    print(summary_df.round(2))
    
    # Create plots for each house
    for house_id in selected_houses:
        fig = analysis.plot_daily_profile(house_id)
        plt.savefig(f'daily_profile_house_{house_id}.png')
        plt.close()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

class SeasonalSolarAnalysis:
    def __init__(self, customer_columns_file, house_load_file, generation_file, ev_load_file):
        """Initialize the analysis with file paths."""
        self.customer_ids = pd.read_csv(customer_columns_file, header=None).iloc[0].values
        self.house_load = pd.read_csv(house_load_file, header=None)
        self.generation = pd.read_csv(generation_file, header=None)
        self.ev_load = pd.read_csv(ev_load_file, header=None)
        
        # Create timestamp index
        self.create_timestamp_index()
        
    def create_timestamp_index(self):
        """Create timestamp index for the data (assuming 30-min intervals)."""
        start_date = datetime(2019, 7, 1)
        timestamps = [start_date + timedelta(minutes=30*i) for i in range(len(self.house_load))]
        
        # Set index for all dataframes
        self.house_load.index = timestamps
        self.generation.index = timestamps
        self.ev_load.index = timestamps
        
    def select_sample_houses(self, n_houses=3):
        """Select a sample of houses for analysis."""
        np.random.seed(0)  # For reproducibility
        # self.selected_houses = np.random.choice(self.customer_ids, size=n_houses, replace=False)
        # 2, 73, 74, 109, 141
        self.selected_houses = np.array([2, 73, 74, 109, 141])
        return self.selected_houses

    def plot_monthly_data(self, house_id, year=2021):
        """Plot complete months of January and July showing daily patterns."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
        
        # Plot January (Summer)
        jan_mask = (house_load.index.year == year) & (house_load.index.month == 1)
        jan_load = house_load[jan_mask].resample('D').sum()
        jan_gen = generation[jan_mask].resample('D').sum()
        
        # Plot January data
        jan_days = range(1, 32)  # January has 31 days
        ax1.plot(jan_days, jan_load.values, 'b-', label='Daily Load', marker='o')
        ax1.plot(jan_days, jan_gen.values, 'r-', label='Daily Generation', marker='o')
        ax1.set_title(f'January (Summer) - Daily Energy - House {house_id}')
        ax1.set_xlabel('Day of Month')
        ax1.set_ylabel('Daily Energy (kWh)')
        ax1.legend()
        ax1.grid(True)
        ax1.set_xticks(jan_days)
        
        # Add self-consumption percentage for January
        daily_self_consumed = np.minimum(jan_load.values, jan_gen.values)
        jan_self_consumption = (daily_self_consumed.sum() / jan_gen.values.sum()) * 100
        ax1.text(0.02, 0.95, f'Monthly Self-Consumption: {jan_self_consumption:.1f}%',
                transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        # Plot July (Winter)
        july_mask = (house_load.index.year == year) & (house_load.index.month == 7)
        july_load = house_load[july_mask].resample('D').sum()
        july_gen = generation[july_mask].resample('D').sum()
        
        # Plot July data
        july_days = range(1, 32)  # July has 31 days
        ax2.plot(july_days, july_load.values, 'b-', label='Daily Load', marker='o')
        ax2.plot(july_days, july_gen.values, 'r-', label='Daily Generation', marker='o')
        ax2.set_title(f'July (Winter) - Daily Energy - House {house_id}')
        ax2.set_xlabel('Day of Month')
        ax2.set_ylabel('Daily Energy (kWh)')
        ax2.legend()
        ax2.grid(True)
        ax2.set_xticks(july_days)
        
        # Add self-consumption percentage for July
        daily_self_consumed = np.minimum(july_load.values, july_gen.values)
        july_self_consumption = (daily_self_consumed.sum() / july_gen.values.sum()) * 100
        ax2.text(0.02, 0.95, f'Monthly Self-Consumption: {july_self_consumption:.1f}%',
                transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return fig
    
    def print_monthly_stats(self, house_id, year=2021):
        """Print monthly statistics for a house."""
        house_load = self.house_load[house_id] #+ self.ev_load[house_id]
        generation = self.generation[house_id]
        
        def calculate_month_stats(month):
            mask = (house_load.index.year == year) & (house_load.index.month == month)
            month_load = house_load[mask].resample('D').sum()
            month_gen = generation[mask].resample('D').sum()
            daily_self_consumed = np.minimum(month_load, month_gen)
            
            return {
                'Total Load (kWh)': month_load.sum(),
                'Total Generation (kWh)': month_gen.sum(),
                'Average Daily Load (kWh)': month_load.mean(),
                'Average Daily Generation (kWh)': month_gen.mean(),
                'Max Daily Load (kWh)': month_load.max(),
                'Max Daily Generation (kWh)': month_gen.max(),
                'Min Daily Load (kWh)': month_load.min(),
                'Min Daily Generation (kWh)': month_gen.min(),
                'Self-Consumption (%)': (daily_self_consumed.sum() / month_gen.sum()) * 100 if month_gen.sum() > 0 else 0,
                'Days with Excess Generation': (month_gen > month_load).sum()
            }
        
        stats = pd.DataFrame({
            'January (Summer)': calculate_month_stats(1),
            'July (Winter)': calculate_month_stats(7)
        })
        
        return stats

def main():
    try:
        # Initialize analysis
        analysis = SeasonalSolarAnalysis(
            'CustomerColumns.csv',
            'HouseLoad.csv',
            'Generation.csv',
            'EVLoad.csv'
        )
        
        # Select sample houses
        selected_houses = analysis.select_sample_houses(5)
        print("Selected houses for analysis:", selected_houses)
        
        # Create seasonal plots and statistics for each house
        for house_id in selected_houses:
            try:
                # Create and save monthly plots
                fig = analysis.plot_monthly_data(house_id)
                plt.savefig(f'monthly_profile_house_{house_id}.png', dpi=300, bbox_inches='tight')
                plt.close()
                
                # Print and save monthly statistics
                stats = analysis.print_monthly_stats(house_id)
                print(f"\nMonthly Statistics for House {house_id}:")
                print(stats.round(2))
                
                # Save statistics to CSV
                stats.to_csv(f'monthly_stats_house_{house_id}.csv')
                
            except Exception as e:
                print(f"Error processing house {house_id}: {str(e)}")
                
    except Exception as e:
        print(f"Error in main execution: {str(e)}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

class ComprehensiveSolarAnalysis:
    def __init__(self, customer_columns_file, house_load_file, generation_file, ev_load_file):
        """Initialize the analysis with file paths."""
        self.customer_ids = pd.read_csv(customer_columns_file, header=None).iloc[0].values
        self.house_load = pd.read_csv(house_load_file, header=None)
        self.generation = pd.read_csv(generation_file, header=None)
        self.ev_load = pd.read_csv(ev_load_file, header=None)
        
        # Create timestamp index
        self.create_timestamp_index()
        
    def create_timestamp_index(self):
        """Create timestamp index for the data (3 years with 30-min intervals)."""
        start_date = datetime(2019, 8, 1)
        timestamps = [start_date + timedelta(minutes=30*i) for i in range(52608)]
        
        # Set index for all dataframes
        self.house_load.index = timestamps
        self.generation.index = timestamps
        self.ev_load.index = timestamps
        
    def select_sample_houses(self, n_houses=5):
        """Select a sample of houses for analysis."""
        np.random.seed(0)  # For reproducibility
        # self.selected_houses = np.random.choice(self.customer_ids, size=n_houses, replace=False)
        self.selected_houses = np.array([2, 73, 74, 109, 141])
        return self.selected_houses

    def calculate_metrics(self, house_id):
        """Calculate self-consumption and self-sufficiency for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate self-consumption
        self_consumed = np.minimum(house_load, generation)
        self_consumption = np.sum(self_consumed) / np.sum(generation) * 100
        
        # Calculate self-sufficiency
        self_sufficiency = np.sum(self_consumed) / np.sum(house_load) * 100
        
        return {
            'house_id': house_id,
            'self_consumption [%]': self_consumption,
            'self_sufficiency [%]': self_sufficiency,
            'total_load [kWh]': np.sum(house_load),
            'total_generation [kWh]': np.sum(generation),
            'total_self_consumed [kWh]': np.sum(self_consumed)
        }

    def plot_daily_profile(self, house_id):
        """Create daily profile plot for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate average daily profile
        daily_load = house_load.groupby(house_load.index.hour).mean()
        daily_generation = generation.groupby(generation.index.hour).mean()
        
        plt.figure(figsize=(10, 6))
        plt.plot(daily_load.index, daily_load.values, label='Consumption (Load + EV)')
        plt.plot(daily_generation.index, daily_generation.values, label='Generation')
        plt.title(f'Average Daily Profile - House {house_id}')
        plt.xlabel('Hour of Day')
        plt.ylabel('Power (kWh)')
        plt.legend()
        plt.grid(True)
        return plt.gcf()

    def plot_daily_average_for_month(self, house_id, year):
        """Create average daily profile plots for January and July in subplots."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
        
        # Plot January (Summer) data
        jan_mask = (house_load.index.year == year) & (house_load.index.month == 1)
        jan_load = house_load[jan_mask].groupby(house_load[jan_mask].index.hour).mean()
        jan_generation = generation[jan_mask].groupby(generation[jan_mask].index.hour).mean()
        
        ax1.plot(jan_load.index, jan_load.values, label='Consumption (Load + EV)')
        ax1.plot(jan_generation.index, jan_generation.values, label='Generation')
        ax1.set_title(f'Summer Average Daily Profile {year} - House {house_id}')
        ax1.set_xlabel('Hour of Day')
        ax1.set_ylabel('Average energy (kWh)')
        ax1.legend()
        ax1.grid(True)
        ax1.set_xticks(range(0, 24))
        
        # Calculate and display January self-consumption
        daily_self_consumed = np.minimum(jan_load, jan_generation)
        jan_self_consumption = (np.sum(daily_self_consumed) / np.sum(jan_generation)) * 100
        ax1.text(0.02, 0.95, f'Average Daily Self-Consumption: {jan_self_consumption:.1f}%',
                transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        # Plot July (Winter) data
        july_mask = (house_load.index.year == year) & (house_load.index.month == 7)
        july_load = house_load[july_mask].groupby(house_load[july_mask].index.hour).mean()
        july_generation = generation[july_mask].groupby(generation[july_mask].index.hour).mean()
        
        ax2.plot(july_load.index, july_load.values, label='Consumption (Load + EV)')
        ax2.plot(july_generation.index, july_generation.values, label='Generation')
        ax2.set_title(f'Winter Average Daily Profile {year} - House {house_id}')
        ax2.set_xlabel('Hour of Day')
        ax2.set_ylabel('Average energy (kWh)')
        ax2.legend()
        ax2.grid(True)
        ax2.set_xticks(range(0, 24))
        
        # Calculate and display July self-consumption
        daily_self_consumed = np.minimum(july_load, july_generation)
        july_self_consumption = (np.sum(daily_self_consumed) / np.sum(july_generation)) * 100
        ax2.text(0.02, 0.95, f'Average Daily Self-Consumption: {july_self_consumption:.1f}%',
                transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return fig

    def plot_monthly_data(self, house_id, year=2021):
        """Plot complete months of January and July showing daily patterns."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
        
        # Plot January (Summer)
        jan_mask = (house_load.index.year == year) & (house_load.index.month == 1)
        jan_load = house_load[jan_mask].resample('D').sum()
        jan_gen = generation[jan_mask].resample('D').sum()
        
        # Plot January data
        jan_days = range(1, 32)  # January has 31 days
        ax1.plot(jan_days, jan_load.values, 'b-', label='Daily consumption (load + EV)')
        ax1.plot(jan_days, jan_gen.values, 'r-', label='Daily generation')
        ax1.set_title(f'January (Summer) - Daily Energy - House {house_id}')
        ax1.set_xlabel('Day of Month')
        ax1.set_ylabel('Energy (kWh)')
        ax1.legend()
        ax1.grid(True)
        ax1.set_xticks(jan_days)
        
        # Add self-consumption percentage for January
        daily_self_consumed = np.minimum(jan_load.values, jan_gen.values)
        jan_self_consumption = (daily_self_consumed.sum() / jan_gen.values.sum()) * 100
        ax1.text(0.02, 0.95, f'Monthly Self-Consumption: {jan_self_consumption:.1f}%',
                transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        # Plot July (Winter)
        july_mask = (house_load.index.year == year) & (house_load.index.month == 7)
        july_load = house_load[july_mask].resample('D').sum()
        july_gen = generation[july_mask].resample('D').sum()
        
        # Plot July data
        july_days = range(1, 32)  # July has 31 days
        ax2.plot(july_days, july_load.values, 'b-', label='Daily consumption (load + EV)')
        ax2.plot(july_days, july_gen.values, 'r-', label='Daily generation')
        ax2.set_title(f'July (Winter) - Daily Energy - House {house_id}')
        ax2.set_xlabel('Day of Month')
        ax2.set_ylabel('Energy (kWh)')
        ax2.legend()
        ax2.grid(True)
        ax2.set_xticks(july_days)
        
        # Add self-consumption percentage for July
        daily_self_consumed = np.minimum(july_load.values, july_gen.values)
        july_self_consumption = (daily_self_consumed.sum() / july_gen.values.sum()) * 100
        ax2.text(0.02, 0.95, f'Monthly Self-Consumption: {july_self_consumption:.1f}%',
                transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return fig

    def print_monthly_stats(self, house_id, year=2021):
        """Print monthly statistics for a house."""
        house_load = self.house_load[house_id] #+ self.ev_load[house_id]
        generation = self.generation[house_id]
        
        def calculate_month_stats(month):
            mask = (house_load.index.year == year) & (house_load.index.month == month)
            month_load = house_load[mask].resample('D').sum()
            month_gen = generation[mask].resample('D').sum()
            daily_self_consumed = np.minimum(month_load, month_gen)
            
            return {
                'Total Load (kWh)': month_load.sum(),
                'Total Generation (kWh)': month_gen.sum(),
                'Average Daily Load (kWh)': month_load.mean(),
                'Average Daily Generation (kWh)': month_gen.mean(),
                'Max Daily Load (kWh)': month_load.max(),
                'Max Daily Generation (kWh)': month_gen.max(),
                'Min Daily Load (kWh)': month_load.min(),
                'Min Daily Generation (kWh)': month_gen.min(),
                'Self-Consumption (%)': (daily_self_consumed.sum() / month_gen.sum()) * 100 if month_gen.sum() > 0 else 0,
                'Days with Excess Generation': (month_gen > month_load).sum()
            }
        
        stats = pd.DataFrame({
            'January (Summer)': calculate_month_stats(1),
            'July (Winter)': calculate_month_stats(7)
        })
        
        return stats

    def create_summary_report(self, selected_houses=None):
        """Create a summary report for selected houses."""
        if selected_houses is None:
            selected_houses = self.selected_houses
            
        # Calculate metrics for all selected houses
        metrics = [self.calculate_metrics(house_id) for house_id in selected_houses]
        return pd.DataFrame(metrics)

def main():
    try:
        # Initialize analysis
        analysis = ComprehensiveSolarAnalysis(
            'CustomerColumns.csv',
            'HouseLoad.csv',
            'Generation.csv',
            'EVLoad.csv'
        )
        
        # Select sample houses
        selected_houses = analysis.select_sample_houses(5)
        print("Selected houses for analysis:", selected_houses)
        
        # Generate summary report
        summary_df = analysis.create_summary_report()
        print("\nSummary Report:")
        print(summary_df.round(2))
        
        # Create plots and statistics for each house
        for house_id in selected_houses:
            try:
                # Create and save daily profile plots
                daily_fig = analysis.plot_daily_profile(house_id)
                plt.savefig(f'daily_profile_house_{house_id}.png')
                plt.close()
                
                # Create and save monthly plots
                monthly_fig = analysis.plot_monthly_data(house_id)
                plt.savefig(f'monthly_profile_house_{house_id}.png', dpi=300, bbox_inches='tight')
                plt.close()

                # Create and save average daily profiles for both months
                daily_avg_fig = analysis.plot_daily_average_for_month(house_id, 2021)
                plt.savefig(f'daily_average_profiles_house_{house_id}.png', dpi=300, bbox_inches='tight')
                plt.close()
                
                # Print and save monthly statistics
                stats = analysis.print_monthly_stats(house_id)
                print(f"\nMonthly Statistics for House {house_id}:")
                print(stats.round(2))
                
                # Save statistics to CSV
                stats.to_csv(f'monthly_stats_house_{house_id}.csv')
                
            except Exception as e:
                print(f"Error processing house {house_id}: {str(e)}")
                
    except Exception as e:
        print(f"Error in main execution: {str(e)}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

class ComprehensiveSolarAnalysis:
    def __init__(self, customer_columns_file, house_load_file, generation_file, ev_load_file):
        """Initialize the analysis with file paths."""
        self.customer_ids = pd.read_csv(customer_columns_file, header=None).iloc[0].values
        self.house_load = pd.read_csv(house_load_file, header=None)
        self.generation = pd.read_csv(generation_file, header=None)
        self.ev_load = pd.read_csv(ev_load_file, header=None)
        
        # Create timestamp index
        self.create_timestamp_index()
        
    def create_timestamp_index(self):
        """Create timestamp index for the data (3 years with 30-min intervals)."""
        start_date = datetime(2019, 8, 1)
        timestamps = [start_date + timedelta(minutes=30*i) for i in range(52608)]
        
        # Set index for all dataframes
        self.house_load.index = timestamps
        self.generation.index = timestamps
        self.ev_load.index = timestamps
        
    def select_sample_houses(self, n_houses=5):
        """Select a sample of houses for analysis."""
        np.random.seed(0)  # For reproducibility
        # self.selected_houses = np.random.choice(self.customer_ids, size=n_houses, replace=False)
        self.selected_houses = np.array([2, 73, 74, 109, 141])
        return self.selected_houses

    def calculate_metrics(self, house_id):
        """Calculate self-consumption and self-sufficiency for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate self-consumption
        self_consumed = np.minimum(house_load, generation)
        self_consumption = np.sum(self_consumed) / np.sum(generation) * 100
        
        # Calculate self-sufficiency
        self_sufficiency = np.sum(self_consumed) / np.sum(house_load) * 100
        
        return {
            'house_id': house_id,
            'self_consumption [%]': self_consumption,
            'self_sufficiency [%]': self_sufficiency,
            'total_load [kWh]': np.sum(house_load),
            'total_generation [kWh]': np.sum(generation),
            'total_self_consumed [kWh]': np.sum(self_consumed)
        }

    def plot_daily_profile(self, house_id):
        """Create daily profile plot for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate average daily profile
        daily_load = house_load.groupby(house_load.index.hour).mean()
        daily_generation = generation.groupby(generation.index.hour).mean()
        
        plt.figure(figsize=(10, 6))
        plt.plot(daily_load.index, daily_load.values, label='Consumption (Load + EV)')
        plt.plot(daily_generation.index, daily_generation.values, label='Generation')
        plt.title(f'Average Daily Profile - House {house_id}')
        plt.xlabel('Hour of Day')
        plt.ylabel('Power (kWh)')
        plt.legend()
        plt.grid(True)
        return plt.gcf()

    def plot_daily_average_for_season(self, house_id, year):
        """Create average daily profile plots for summer (Dec-Feb) and winter (Jun-Aug) in subplots."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
        
        # Plot Summer data (Dec-Feb)
        summer_months = [12, 1, 2]  # December, January, February
        summer_data = pd.DataFrame()
        
        for month in summer_months:
            year_to_use = year if month != 12 else year - 1  # Use previous year for December
            mask = (house_load.index.year == year_to_use) & (house_load.index.month == month)
            
            # Group by hour and get mean for this month
            month_load = house_load[mask].groupby(house_load[mask].index.hour).mean()
            month_gen = generation[mask].groupby(generation[mask].index.hour).mean()
            
            if summer_data.empty:
                summer_data['load'] = month_load
                summer_data['generation'] = month_gen
            else:
                summer_data['load'] += month_load
                summer_data['generation'] += month_gen
        
        # Calculate averages across the three months
        summer_data = summer_data / len(summer_months)
        
        ax1.plot(summer_data.index, summer_data['load'], label='Consumption (Load + EV)')
        ax1.plot(summer_data.index, summer_data['generation'], label='Generation')
        ax1.set_title(f'Summer (Dec-Feb) Average Daily Profile {year} - House {house_id}')
        ax1.set_xlabel('Hour of Day')
        ax1.set_ylabel('Average energy (kWh)')
        ax1.legend()
        ax1.grid(True)
        ax1.set_xticks(range(0, 24))
        
        # Calculate and display summer self-consumption
        daily_self_consumed = np.minimum(summer_data['load'], summer_data['generation'])
        summer_self_consumption = (np.sum(daily_self_consumed) / np.sum(summer_data['generation'])) * 100
        ax1.text(0.02, 0.95, f'Average Daily Self-Consumption: {summer_self_consumption:.1f}%',
                transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        # Plot Winter data (Jun-Aug)
        winter_months = [6, 7, 8]  # June, July, August
        winter_data = pd.DataFrame()
        
        for month in winter_months:
            mask = (house_load.index.year == year) & (house_load.index.month == month)
            
            # Group by hour and get mean for this month
            month_load = house_load[mask].groupby(house_load[mask].index.hour).mean()
            month_gen = generation[mask].groupby(generation[mask].index.hour).mean()
            
            if winter_data.empty:
                winter_data['load'] = month_load
                winter_data['generation'] = month_gen
            else:
                winter_data['load'] += month_load
                winter_data['generation'] += month_gen
        
        # Calculate averages across the three months
        winter_data = winter_data / len(winter_months)
        
        ax2.plot(winter_data.index, winter_data['load'], label='Consumption (Load + EV)')
        ax2.plot(winter_data.index, winter_data['generation'], label='Generation')
        ax2.set_title(f'Winter (Jun-Aug) Average Daily Profile {year} - House {house_id}')
        ax2.set_xlabel('Hour of Day')
        ax2.set_ylabel('Average energy (kWh)')
        ax2.legend()
        ax2.grid(True)
        ax2.set_xticks(range(0, 24))
        
        # Calculate and display winter self-consumption
        daily_self_consumed = np.minimum(winter_data['load'], winter_data['generation'])
        winter_self_consumption = (np.sum(daily_self_consumed) / np.sum(winter_data['generation'])) * 100
        ax2.text(0.02, 0.95, f'Average Daily Self-Consumption: {winter_self_consumption:.1f}%',
                transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return fig

    def plot_seasonal_data(self, house_id, year=2021):
        """Plot seasonal averages (Summer: Dec-Feb, Winter: Jun-Aug) showing daily patterns."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
        
        # Process Summer data (Dec-Feb)
        summer_months = [12, 1, 2]  # December, January, February
        summer_data = pd.DataFrame()
        
        for month in summer_months:
            year_to_use = year if month != 12 else year - 1  # Use previous year for December
            mask = (house_load.index.year == year_to_use) & (house_load.index.month == month)
            
            # Resample to daily values for this month
            month_load = house_load[mask].resample('D').sum()
            month_gen = generation[mask].resample('D').sum()
            
            if summer_data.empty:
                summer_data['load'] = month_load
                summer_data['generation'] = month_gen
            else:
                summer_data['load'] = summer_data['load'].add(month_load, fill_value=0)
                summer_data['generation'] = summer_data['generation'].add(month_gen, fill_value=0)
        
        # Calculate averages across the three months
        summer_data = summer_data / len(summer_months)
        
        # Plot Summer data
        days = range(len(summer_data))
        ax1.plot(days, summer_data['load'], 'b-', label='Daily consumption (load + EV)')
        ax1.plot(days, summer_data['generation'], label='Daily generation')
        ax1.set_title(f'Summer (Dec-Feb) - Average Daily Energy - House {house_id}')
        ax1.set_xlabel('Days in Season')
        ax1.set_ylabel('Average Energy (kWh)')
        ax1.legend()
        ax1.grid(True)
        
        # Add self-consumption percentage for Summer
        daily_self_consumed = np.minimum(summer_data['load'], summer_data['generation'])
        summer_self_consumption = (daily_self_consumed.sum() / summer_data['generation'].sum()) * 100
        ax1.text(0.02, 0.95, f'Average Seasonal Self-Consumption: {summer_self_consumption:.1f}%',
                transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        # Process Winter data (Jun-Aug)
        winter_months = [6, 7, 8]  # June, July, August
        winter_data = pd.DataFrame()
        
        for month in winter_months:
            mask = (house_load.index.year == year) & (house_load.index.month == month)
            
            # Resample to daily values for this month
            month_load = house_load[mask].resample('D').sum()
            month_gen = generation[mask].resample('D').sum()
            
            if winter_data.empty:
                winter_data['load'] = month_load
                winter_data['generation'] = month_gen
            else:
                winter_data['load'] = winter_data['load'].add(month_load, fill_value=0)
                winter_data['generation'] = winter_data['generation'].add(month_gen, fill_value=0)
        
        # Calculate averages across the three months
        winter_data = winter_data / len(winter_months)
        
        # Plot Winter data
        days = range(len(winter_data))
        ax2.plot(days, winter_data['load'], label='Daily consumption (load + EV)')
        ax2.plot(days, winter_data['generation'], label='Daily generation')
        ax2.set_title(f'Winter - Average Daily Energy - House {house_id}')
        ax2.set_xlabel('Days in Season')
        ax2.set_ylabel('Average Energy (kWh)')
        ax2.legend()
        ax2.grid(True)
        
        # Add self-consumption percentage for Winter
        daily_self_consumed = np.minimum(winter_data['load'], winter_data['generation'])
        winter_self_consumption = (daily_self_consumed.sum() / winter_data['generation'].sum()) * 100
        ax2.text(0.02, 0.95, f'Average Seasonal Self-Consumption: {winter_self_consumption:.1f}%',
                transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return fig

    def print_monthly_stats(self, house_id, year=2021):
        """Print monthly statistics for a house."""
        house_load = self.house_load[house_id] #+ self.ev_load[house_id]
        generation = self.generation[house_id]
        
        def calculate_month_stats(month):
            mask = (house_load.index.year == year) & (house_load.index.month == month)
            month_load = house_load[mask].resample('D').sum()
            month_gen = generation[mask].resample('D').sum()
            daily_self_consumed = np.minimum(month_load, month_gen)
            
            return {
                'Total Load (kWh)': month_load.sum(),
                'Total Generation (kWh)': month_gen.sum(),
                'Average Daily Load (kWh)': month_load.mean(),
                'Average Daily Generation (kWh)': month_gen.mean(),
                'Max Daily Load (kWh)': month_load.max(),
                'Max Daily Generation (kWh)': month_gen.max(),
                'Min Daily Load (kWh)': month_load.min(),
                'Min Daily Generation (kWh)': month_gen.min(),
                'Self-Consumption (%)': (daily_self_consumed.sum() / month_gen.sum()) * 100 if month_gen.sum() > 0 else 0,
                'Days with Excess Generation': (month_gen > month_load).sum()
            }
        
        stats = pd.DataFrame({
            'January (Summer)': calculate_month_stats(1),
            'July (Winter)': calculate_month_stats(7)
        })
        
        return stats

    def create_summary_report(self, selected_houses=None):
        """Create a summary report for selected houses."""
        if selected_houses is None:
            selected_houses = self.selected_houses
            
        # Calculate metrics for all selected houses
        metrics = [self.calculate_metrics(house_id) for house_id in selected_houses]
        return pd.DataFrame(metrics)

def main():
    try:
        # Initialize analysis
        analysis = ComprehensiveSolarAnalysis(
            'CustomerColumns.csv',
            'HouseLoad.csv',
            'Generation.csv',
            'EVLoad.csv'
        )
        
        # Select sample houses
        selected_houses = analysis.select_sample_houses(5)
        print("Selected houses for analysis:", selected_houses)
        
        # Generate summary report
        summary_df = analysis.create_summary_report()
        print("\nSummary Report:")
        print(summary_df.round(2))
        
        # Create plots and statistics for each house
        for house_id in selected_houses:
            try:
                # Create and save daily profile plots
                daily_fig = analysis.plot_daily_profile(house_id)
                plt.savefig(f'daily_profile_house_{house_id}.png')
                plt.close()
                
                # Create and save monthly plots
                monthly_fig = analysis.plot_seasonal_data(house_id)
                plt.savefig(f'monthly_profile_house_{house_id}.png', dpi=300, bbox_inches='tight')
                plt.close()

                # Create and save average daily profiles for both months
                daily_avg_fig = analysis.plot_daily_average_for_season(house_id, 2021)
                plt.savefig(f'daily_average_profiles_house_{house_id}.png', dpi=300, bbox_inches='tight')
                plt.close()
                
                # Print and save monthly statistics
                stats = analysis.print_monthly_stats(house_id)
                print(f"\nMonthly Statistics for House {house_id}:")
                print(stats.round(2))
                
                # Save statistics to CSV
                stats.to_csv(f'monthly_stats_house_{house_id}.csv')
                
            except Exception as e:
                print(f"Error processing house {house_id}: {str(e)}")
                
    except Exception as e:
        print(f"Error in main execution: {str(e)}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

class ComprehensiveSolarAnalysis:
    def __init__(self, customer_columns_file, house_load_file, generation_file, ev_load_file):
        """Initialize the analysis with file paths."""
        self.customer_ids = pd.read_csv(customer_columns_file, header=None).iloc[0].values
        self.house_load = pd.read_csv(house_load_file, header=None)
        self.generation = pd.read_csv(generation_file, header=None)
        self.ev_load = pd.read_csv(ev_load_file, header=None)
        
        # Create timestamp index
        self.create_timestamp_index()
        
    def create_timestamp_index(self):
        """Create timestamp index for the data (3 years with 30-min intervals)."""
        start_date = datetime(2019, 8, 1)
        timestamps = [start_date + timedelta(minutes=30*i) for i in range(52608)]
        
        # Set index for all dataframes
        self.house_load.index = timestamps
        self.generation.index = timestamps
        self.ev_load.index = timestamps
        
    def select_sample_houses(self, n_houses=5):
        """Select a sample of houses for analysis."""
        np.random.seed(0)  # For reproducibility
        # self.selected_houses = np.random.choice(self.customer_ids, size=n_houses, replace=False)
        self.selected_houses = np.array([2, 73, 74, 109, 141])
        return self.selected_houses

    def calculate_metrics(self, house_id):
        """Calculate self-consumption and self-sufficiency for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate self-consumption
        self_consumed = np.minimum(house_load, generation)
        self_consumption = np.sum(self_consumed) / np.sum(generation) * 100
        
        # Calculate self-sufficiency
        self_sufficiency = np.sum(self_consumed) / np.sum(house_load) * 100
        
        return {
            'house_id': house_id,
            'self_consumption [%]': self_consumption,
            'self_sufficiency [%]': self_sufficiency,
            'total_load [kWh]': np.sum(house_load),
            'total_generation [kWh]': np.sum(generation),
            'total_self_consumed [kWh]': np.sum(self_consumed)
        }

    def plot_daily_profile(self, house_id):
        """Create daily profile plot for a specific house."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Calculate average daily profile
        daily_load = house_load.groupby(house_load.index.hour).mean()
        daily_generation = generation.groupby(generation.index.hour).mean()
        
        plt.figure(figsize=(10, 6))
        plt.plot(daily_load.index, daily_load.values, label='Consumption (Load + EV)')
        plt.plot(daily_generation.index, daily_generation.values, label='Generation')
        plt.title(f'Average Daily Profile - House {house_id}')
        plt.xlabel('Hour of Day')
        plt.ylabel('Power (kWh)')
        plt.legend()
        plt.grid(True)
        return plt.gcf()

    def plot_daily_average_for_season(self, house_id, year):
        """Create average daily profile plots for summer (Dec-Feb) and winter (Jun-Aug) in subplots."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
        
        # Plot Summer data (Dec-Feb)
        summer_months = [12, 1, 2]  # December, January, February
        summer_data = pd.DataFrame()
        
        for month in summer_months:
            year_to_use = year if month != 12 else year - 1  # Use previous year for December
            mask = (house_load.index.year == year_to_use) & (house_load.index.month == month)
            
            # Group by hour and get mean for this month
            month_load = house_load[mask].groupby(house_load[mask].index.hour).mean()
            month_gen = generation[mask].groupby(generation[mask].index.hour).mean()
            
            if summer_data.empty:
                summer_data['load'] = month_load
                summer_data['generation'] = month_gen
            else:
                summer_data['load'] += month_load
                summer_data['generation'] += month_gen
        
        # Calculate averages across the three months
        summer_data = summer_data / len(summer_months)
        
        ax1.plot(summer_data.index, summer_data['load'], label='Consumption (Load + EV)')
        ax1.plot(summer_data.index, summer_data['generation'], label='Generation')
        ax1.set_title(f'Summer (Dec-Feb) Average Daily Profile {year} - House {house_id}')
        ax1.set_xlabel('Hour of Day')
        ax1.set_ylabel('Average energy (kWh)')
        ax1.legend()
        ax1.grid(True)
        ax1.set_xticks(range(0, 24))
        
        # Calculate and display summer self-consumption
        daily_self_consumed = np.minimum(summer_data['load'], summer_data['generation'])
        summer_self_consumption = (np.sum(daily_self_consumed) / np.sum(summer_data['generation'])) * 100
        ax1.text(0.02, 0.95, f'Average Daily Self-Consumption: {summer_self_consumption:.1f}%',
                transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        # Plot Winter data (Jun-Aug)
        winter_months = [6, 7, 8]  # June, July, August
        winter_data = pd.DataFrame()
        
        for month in winter_months:
            mask = (house_load.index.year == year) & (house_load.index.month == month)
            
            # Group by hour and get mean for this month
            month_load = house_load[mask].groupby(house_load[mask].index.hour).mean()
            month_gen = generation[mask].groupby(generation[mask].index.hour).mean()
            
            if winter_data.empty:
                winter_data['load'] = month_load
                winter_data['generation'] = month_gen
            else:
                winter_data['load'] += month_load
                winter_data['generation'] += month_gen
        
        # Calculate averages across the three months
        winter_data = winter_data / len(winter_months)
        
        ax2.plot(winter_data.index, winter_data['load'], label='Consumption (Load + EV)')
        ax2.plot(winter_data.index, winter_data['generation'], label='Generation')
        ax2.set_title(f'Winter (Jun-Aug) Average Daily Profile {year} - House {house_id}')
        ax2.set_xlabel('Hour of Day')
        ax2.set_ylabel('Average energy (kWh)')
        ax2.legend()
        ax2.grid(True)
        ax2.set_xticks(range(0, 24))
        
        # Calculate and display winter self-consumption
        daily_self_consumed = np.minimum(winter_data['load'], winter_data['generation'])
        winter_self_consumption = (np.sum(daily_self_consumed) / np.sum(winter_data['generation'])) * 100
        ax2.text(0.02, 0.95, f'Average Daily Self-Consumption: {winter_self_consumption:.1f}%',
                transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return fig

    def plot_seasonal_data(self, house_id, year=2021):
        """Plot seasonal averages (Summer: Dec-Feb, Winter: Jun-Aug) showing daily patterns."""
        # Get data for specific house
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
        
        # Process Summer data (Dec-Feb)
        summer_months = [12, 1, 2]  # December, January, February
        summer_data = pd.DataFrame()
        
        for month in summer_months:
            year_to_use = year if month != 12 else year - 1  # Use previous year for December
            mask = (house_load.index.year == year_to_use) & (house_load.index.month == month)
            
            # Resample to daily values for this month
            month_load = house_load[mask].resample('D').sum()
            month_gen = generation[mask].resample('D').sum()
            
            if summer_data.empty:
                summer_data['load'] = month_load
                summer_data['generation'] = month_gen
            else:
                summer_data['load'] = summer_data['load'].add(month_load, fill_value=0)
                summer_data['generation'] = summer_data['generation'].add(month_gen, fill_value=0)
        
        # Calculate averages across the three months
        summer_data = summer_data / len(summer_months)
        
        # Plot Summer data
        days = range(len(summer_data))
        ax1.plot(days, summer_data['load'], 'b-', label='Daily consumption (load + EV)')
        ax1.plot(days, summer_data['generation'], label='Daily generation')
        ax1.set_title(f'Summer (Dec-Feb) - Average Daily Energy - House {house_id}')
        ax1.set_xlabel('Days in Season')
        ax1.set_ylabel('Average Energy (kWh)')
        ax1.legend()
        ax1.grid(True)
        
        # Add self-consumption percentage for Summer
        daily_self_consumed = np.minimum(summer_data['load'], summer_data['generation'])
        summer_self_consumption = (daily_self_consumed.sum() / summer_data['generation'].sum()) * 100
        ax1.text(0.02, 0.95, f'Average Seasonal Self-Consumption: {summer_self_consumption:.1f}%',
                transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        # Process Winter data (Jun-Aug)
        winter_months = [6, 7, 8]  # June, July, August
        winter_data = pd.DataFrame()
        
        for month in winter_months:
            mask = (house_load.index.year == year) & (house_load.index.month == month)
            
            # Resample to daily values for this month
            month_load = house_load[mask].resample('D').sum()
            month_gen = generation[mask].resample('D').sum()
            
            if winter_data.empty:
                winter_data['load'] = month_load
                winter_data['generation'] = month_gen
            else:
                winter_data['load'] = winter_data['load'].add(month_load, fill_value=0)
                winter_data['generation'] = winter_data['generation'].add(month_gen, fill_value=0)
        
        # Calculate averages across the three months
        winter_data = winter_data / len(winter_months)
        
        # Plot Winter data
        days = range(len(winter_data))
        ax2.plot(days, winter_data['load'], label='Daily consumption (load + EV)')
        ax2.plot(days, winter_data['generation'], label='Daily generation')
        ax2.set_title(f'Winter - Average Daily Energy - House {house_id}')
        ax2.set_xlabel('Days in Season')
        ax2.set_ylabel('Average Energy (kWh)')
        ax2.legend()
        ax2.grid(True)
        
        # Add self-consumption percentage for Winter
        daily_self_consumed = np.minimum(winter_data['load'], winter_data['generation'])
        winter_self_consumption = (daily_self_consumed.sum() / winter_data['generation'].sum()) * 100
        ax2.text(0.02, 0.95, f'Average Seasonal Self-Consumption: {winter_self_consumption:.1f}%',
                transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return fig

    def create_summary_report(self, selected_houses=None):
        """Create a summary report for selected houses showing annual metrics."""
        if selected_houses is None:
            selected_houses = self.selected_houses
        
        metrics_list = []
        for house_id in selected_houses:
            # Get data for specific house
            house_load = self.house_load[house_id] + self.ev_load[house_id]
            generation = self.generation[house_id]
            
            # Calculate self-consumption and self-sufficiency for entire year
            self_consumed = np.minimum(house_load, generation)
            self_consumption = np.sum(self_consumed) / np.sum(generation) * 100
            self_sufficiency = np.sum(self_consumed) / np.sum(house_load) * 100
            
            metrics = {
                'house_id': house_id,
                'self_consumption [%]': self_consumption,
                'self_sufficiency [%]': self_sufficiency,
                'total_load [kWh]': np.sum(house_load),
                'total_generation [kWh]': np.sum(generation),
                'total_self_consumed [kWh]': np.sum(self_consumed)
            }
            metrics_list.append(metrics)
        
        return pd.DataFrame(metrics_list)

    def print_seasonal_stats(self, house_id, year=2021):
        """Print seasonal statistics for summer (Dec-Feb) and winter (Jun-Aug)."""
        house_load = self.house_load[house_id] + self.ev_load[house_id]
        generation = self.generation[house_id]
        
        def calculate_seasonal_stats(season_months, season_year):
            """Calculate statistics for a given season."""
            seasonal_data = {'load': [], 'generation': []}
            
            for month in season_months:
                year_to_use = season_year if month != 12 else season_year - 1
                mask = (house_load.index.year == year_to_use) & (house_load.index.month == month)
                month_load = house_load[mask].resample('D').sum()
                month_gen = generation[mask].resample('D').sum()
                
                seasonal_data['load'].extend(month_load.values)
                seasonal_data['generation'].extend(month_gen.values)
            
            # Convert to numpy arrays for calculations
            seasonal_load = np.array(seasonal_data['load'])
            seasonal_gen = np.array(seasonal_data['generation'])
            daily_self_consumed = np.minimum(seasonal_load, seasonal_gen)
            
            return {
                'Total Load (kWh)': np.sum(seasonal_load),
                'Total Generation (kWh)': np.sum(seasonal_gen),
                'Average Daily Load (kWh)': np.mean(seasonal_load),
                'Average Daily Generation (kWh)': np.mean(seasonal_gen),
                'Max Daily Load (kWh)': np.max(seasonal_load),
                'Max Daily Generation (kWh)': np.max(seasonal_gen),
                'Min Daily Load (kWh)': np.min(seasonal_load),
                'Min Daily Generation (kWh)': np.min(seasonal_gen),
                'Self-Consumption (%)': (np.sum(daily_self_consumed) / np.sum(seasonal_gen)) * 100 if np.sum(seasonal_gen) > 0 else 0,
                'Days with Excess Generation': np.sum(seasonal_gen > seasonal_load)
            }
        
        # Calculate statistics for summer (Dec-Feb) and winter (Jun-Aug)
        stats = pd.DataFrame({
            'Summer (Dec-Feb)': calculate_seasonal_stats([12, 1, 2], year),
            'Winter (Jun-Aug)': calculate_seasonal_stats([6, 7, 8], year)
        })
        
        return stats

def main():
    try:
        # Initialize analysis
        print("\nInitializing Solar Analysis...")
        analysis = ComprehensiveSolarAnalysis(
            'CustomerColumns.csv',
            'HouseLoad.csv',
            'Generation.csv',
            'EVLoad.csv'
        )
        
        # Select sample houses
        selected_houses = analysis.select_sample_houses(5)
        print("\nSelected houses for analysis:", selected_houses)
        
        # Generate and display annual summary report
        print("\nGenerating Annual Summary Report...")
        summary_df = analysis.create_summary_report()
        print("\nAnnual Summary Report:")
        print(summary_df.round(2))
        
        # Save annual summary report
        summary_df.to_csv('annual_summary_report.csv', index=False)
        print("\nAnnual summary report saved to 'annual_summary_report.csv'")
        
        # Process each house
        print("\nGenerating detailed analysis for each house...")
        for house_id in selected_houses:
            try:
                print(f"\nProcessing House {house_id}...")
                
                # Create and save daily profile plots
                print(f"- Creating daily profile plot...")
                daily_fig = analysis.plot_daily_profile(house_id)
                plt.savefig(f'house_{house_id}_daily_profile.png', dpi=300, bbox_inches='tight')
                plt.close()
                
                # Create and save seasonal average plots
                print(f"- Creating seasonal average plots...")
                seasonal_avg_fig = analysis.plot_daily_average_for_season(house_id, 2021)
                plt.savefig(f'house_{house_id}_seasonal_daily_averages.png', dpi=300, bbox_inches='tight')
                plt.close()

                # Create and save seasonal profile plots
                print(f"- Creating seasonal profile plots...")
                seasonal_fig = analysis.plot_seasonal_data(house_id)
                plt.savefig(f'house_{house_id}_seasonal_profiles.png', dpi=300, bbox_inches='tight')
                plt.close()
                
                # Generate and save seasonal statistics
                print(f"- Calculating seasonal statistics...")
                seasonal_stats = analysis.print_seasonal_stats(house_id)
                
                print(f"\nSeasonal Statistics for House {house_id}:")
                print(seasonal_stats.round(2))
                
                # Save seasonal statistics to CSV
                stats_filename = f'house_{house_id}_seasonal_stats.csv'
                seasonal_stats.to_csv(stats_filename)
                print(f"Seasonal statistics saved to '{stats_filename}'")
                
                print(f"Successfully completed analysis for House {house_id}")
                
            except Exception as e:
                print(f"Error processing house {house_id}: {str(e)}")
                print("Continuing with next house...")
                continue
        
        print("\nAnalysis complete! All files have been saved.")
        
    except FileNotFoundError as e:
        print(f"\nError: Required input file not found: {str(e)}")
        print("Please ensure all required CSV files are in the correct directory:")
        print("- CustomerColumns.csv")
        print("- HouseLoad.csv")
        print("- Generation.csv")
        print("- EVLoad.csv")
    except Exception as e:
        print(f"\nError in main execution: {str(e)}")
    finally:
        print("\nProgram execution finished.")

if __name__ == "__main__":
    main()
```
